```python
from sklearn.svm import SVC
import matplotlib.pyplot as plt
from sklearn.datasets import make_circles, make_moons
from mlxtend.plotting import plot_decision_regions

import warnings
warnings.filterwarnings('ignore')
```

```python
x, y = make_circles(100, factor=0.1, noise=0.1)
plt.scatter(x[:,0], x[:,1], c=y)
```

```python
linear_svm = SVC(kernel='linear').fit(x,y)
plot_decision_regions(x, y, clf=linear_svm)
```

```python
svm = SVC(kernel='rbf')
svm.fit(x,y)

plot_decision_regions(x,y,clf=svm)
```

Сгенерируем две лунки, точки которых представляют разные классы в задаче бинарной классификации

```python
x,y = make_moons(100, noise=0.1)
plt.scatter(x[:,0],x[:,1], c=y)
```

```python
linear_svm = SVC(kernel='linear').fit(x,y)
plot_decision_regions(x, y, clf=linear_svm)
```

```python
svm = SVC(kernel='rbf', C=1000).fit(x,y)
plot_decision_regions(x,y, clf=svm)
```

```python
svm_poly = SVC(kernel='poly', degree=3).fit(x,y)
plot_decision_regions(x, y, svm_poly)
```
